##Data preparation üìá <a class="anchor" id="prep"></a>

In this section we are going to preprocess our data.

First, we want to split the data in X and y, where y will have FLAG feature and X the rest of features.

```python
y = filtered_transactions.FLAG
X = filtered_transactions.drop(columns=['FLAG']) 
```

We have to split between train and test sets. We are going to have 80% for train and 20% for test.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 1)
```

As we've seen before, our data is not normalized and is not balanced so we are going to fix this.

Let's start normalizing our numerical features.

```python
# set normalizer
normalizer = Normalizer()

# normalizing train
norm_train = normalizer.fit_transform(X_train[num_col].values)
norm_train_df = pd.DataFrame(norm_train, index=X_train[num_col].index, columns=X_train[num_col].columns)

# normalizing test
norm_test = normalizer.fit_transform(X_test[num_col].values)
norm_test_df = pd.DataFrame(norm_test, index=X_test[num_col].index, columns=X_test[num_col].columns)
```

We have to get dummy variables from our categorical columns to pass them into the model. We are going to do that using one hot encoding.

```python
cat_cols_encoded = []
for col in cat_col:
  cat_cols_encoded += [f"{col[0]}_{cat}" for cat in list(X_train[col].unique())]

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_cols_train = encoder.fit_transform(X_train[cat_col])
enc_train_df = pd.DataFrame(encoded_cols_train, columns=cat_cols_encoded, index=X_train[cat_col].index)
enc_train_df = enc_train_df.loc[:,~enc_train_df.columns.duplicated()].copy()

encoded_cols_test = encoder.transform(X_test[cat_col])
enc_test_df = pd.DataFrame(encoded_cols_test, columns=cat_cols_encoded, index=X_test[cat_col].index)
enc_test_df = enc_test_df.loc[:,~enc_test_df.columns.duplicated()].copy()
```

Now we concat categorical and numerical columns.

```python
X_train_norm = pd.concat([norm_train_df, enc_train_df], axis=1)
X_test_norm = pd.concat([norm_test_df, enc_test_df], axis=1)
```

We are going to solve the data imbalaced issue using SMOTE. For th

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=42)

X_train_final, y_train_final = sm.fit_resample(X_train_norm, y_train)
X_test_final, y_test_final = sm.fit_resample(X_test_norm, y_test)
```

Let's check that the data now is balanced.

```python
print('TRAIN')
print('Percentage flagged:', np.count_nonzero(y_train_final == 1)/len(y_train_final))
print('Percentage not flagged:', np.count_nonzero(y_train_final == 0)/len(y_train_final))
print()
print('TEST')
print('Percentage flagged:', np.count_nonzero(y_test_final == 1)/len(y_test_final))
print('Percentage not flagged:', np.count_nonzero(y_test_final == 0)/len(y_test_final))
```

##Classification modelüéØ <a class="anchor" id="model"></a>
We are going to develop a classification model with XGboost, using the features selected in the previous sections. 

Let's initialize the model selecting some hyperparameters.

```python
model = XGBClassifier(
    objective= 'binary:logistic',
    nthread=4,
    seed=42,
    max_depth= 10,
    reg_alpha= 0.5,
    reg_lambda= 0.5,
)
```

Now we are goint to fit the model.

```python
model.fit(X_train_final, y_train_final)
```

First, we are checking accuracy, getting 97.84% accuracy which is a really good value. Remember that accuracy goes to 0% to 100%, being the closer to 100% the better.
